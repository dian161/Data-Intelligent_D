````python
# membuat df dgn daftar anggota
con2=pd.DataFrame({"A":1.,
                   "B": pd.Timestamp("20180925"),
                   "C": np.array([3]*4, dtype="int32"),
                   "D": pd.Categorical(["test", "train"]*2),
                   "E": np.array([1,12,32,14])}, index=["satu","dua","tiga","empat"])
con2
````

# 2 Menambah, delete, dan menyeleksi kolom dataframe

hapus data kolom dgn fungsi:
1. del(data_frame["kolom])
2. .pop("kolom)
3. .drop("kolom", axis=1): axis=1 <<menghapus sepanjang kolom yg dipilih

````python
# add kolom baru
con2["F"]=con2["C"]+con2["E"]
print(con2)
````

````python
# add kolomm baru pada index 1,3, dan 4
con2["G"]=pd.Series([10,20,30], index=["satu","tiga","empat"])
print(con2)
````

````python
# delete kolom B
del(con2["B"])
con2
````

````python
# hapus kolom "C" dgn pop
con2.pop("C")
con2
````

````python
# hapus dengan drop
con2.drop("G", axis=1)
````

````python
# nyeleksi kolom "A"
kol_a=con2["A"]
kol_a
````

# 3 Menambah, mengurang, dan menyeleksi baris pd df

```
1. loc = seleksi indeks/bil bulat
2. iloc = seleksi menggunakan bil bulat

tambah baris dengan fungsi = .append()
hapus baris dengan fungsi = .del("index", axis=0)
```

````python
# menambahkan baris baru
con3=pd.DataFrame([[1, "train", 4,5,6.0], [2,"test", 7,12,34]], columns=["A","D","E","F","G"])
con3
````

````python
con4=con2.append(con3)
con4
````

````python
# menyeleksi indeks "tiga" pd baris
con4.loc["tiga"]
````

````python
# menyeleksi baris ke 2 dan 3 dgn menyebut urutan kemudian copy hasilnya
con5=con4.iloc[1:3].copy()
con5
````

````python
# menghapus baris dgn indeks "tiga" simpan dgn var baru
con5_del=con5.drop("tiga", axis=0)
con5_del
````
